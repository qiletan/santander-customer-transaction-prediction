```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import lightgbm as lgb
import os
import shutil
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from tqdm import tqdm
init_notebook_mode(connected=True)

pd.options.display.max_columns = 500
#import pandas_profiling

def display_importances(feature_importance_df_):
    cols = feature_importance_df_[["feature", "importance"]].groupby("feature").mean().sort_values(by="importance", ascending=False)[:50].index
    best_features = feature_importance_df_.loc[feature_importance_df_.feature.isin(cols)]
    plt.figure(figsize=(8, 10))
    sns.barplot(x="importance", y="feature", data=best_features.sort_values(by="importance", ascending=False))
    plt.title('LightGBM Features (avg over folds)')
    plt.tight_layout()
    #plt.savefig('lgbm_importances01.png')
```

```python
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')

var_list = [f for f in train_df.columns if 'var' in f]
```

```python
bin_cols = [f + '_bin' for f in var_list]
```

```python
train_df[bin_cols] = train_df[var_list].apply(lambda x: pd.cut(x,bins=200,labels=False))
```

```python
target_cols = ['te_' + f + '_bin' for f in var_list]
```

```python
for var in bin_cols:
    train_df[f'te_{var}'] = train_df.groupby(var)['target'].transform('mean')
```

```python
pos_sample = train_df.loc[train_df.target==1,target_cols].sample(5)
```

```python
neg_sample = train_df.loc[train_df.target==0,target_cols].sample(5)
```

```python
train_df['sum_te'] = train_df[target_cols].sum(1)
```

```python
train_df.sort_values(['target','sum_te'],ascending=[True,False]).head()
```

```python
most_correct_neg = train_df.sort_values(['target','sum_te'],ascending=[True,True]).head(100).sample()
most_correct_pos = train_df.sort_values(['target','sum_te'],ascending=[False,False]).head(100).sample()



trace = go.Scatter(
    x = [i for i in range(200)],
    y = most_correct_neg[target_cols].values[0],
    mode = 'lines+markers',name='most_correct_neg'
)
trace1 = go.Scatter(
    x = [i for i in range(200)],
    y = most_correct_pos[target_cols].values[0],
    mode = 'lines+markers',name='most_correct_pos'
)

data = [trace,trace1]

# Plot and embed in ipython notebook!
iplot(data)
```

```python
most_wrong_neg = train_df.sort_values(['target','sum_te'],ascending=[True,False]).head(100).sample()
most_wrong_pos = train_df.sort_values(['target','sum_te'],ascending=[False,True]).head(100).sample()



trace = go.Scatter(
    x = [i for i in range(200)],
    y = most_wrong_neg[target_cols].values[0],
    mode = 'lines+markers',name='most_wrong_neg'
)
trace1 = go.Scatter(
    x = [i for i in range(200)],
    y = most_wrong_pos[target_cols].values[0],
    mode = 'lines+markers',name='most_wrong_pos'
)

data = [trace,trace1]

# Plot and embed in ipython notebook!
iplot(data)
```
